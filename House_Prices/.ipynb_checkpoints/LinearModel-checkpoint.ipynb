House Prices - Advanced Regression Techniques from Kaggle

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import root_mean_squared_error
```

```python
X_train = pd.read_csv('data/train.csv')
X_test = pd.read_csv('data/test.csv')
y_train = X_train['SalePrice']

X_train.drop(['Id', 'SalePrice'], axis=1, inplace=True)
X_test.drop('Id', axis=1, inplace=True)
X_train.shape
```

```python
pd.set_option('display.max_columns', None)
X_train.isnull().sum().to_frame().T
```

Fill Nan object values

```python
cat_cols = X_train.select_dtypes(include='object').columns
X_train[cat_cols] = X_train[cat_cols].fillna("0")
X_test[cat_cols] = X_test[cat_cols].fillna("0")
X_train.isnull().sum().to_frame().T
```

Fill Nan numerical values

```python
num_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
X_train[num_cols] = X_train[num_cols].fillna(X_train[num_cols].mean())
X_test[num_cols] = X_test[num_cols].fillna(X_test[num_cols].mean())
print(X_test.isnull().sum().sum())
X_train.isnull().sum().sum()
```

0 Nan values in both Train and Test

```python
X_train = pd.get_dummies(X_train, columns=cat_cols, drop_first=True)
X_test = pd.get_dummies(X_test, columns=cat_cols, drop_first=True)

X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
X_train.dtypes.to_frame().T
```

```python
LR = LinearRegression().fit(X_train,y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

y_pred = LR.predict(X_train)
y_pred = np.clip(y_pred, a_min=1, a_max=None)
rmse_log = np.sqrt(mean_squared_error(np.log(y_train), np.log(y_pred)))
print("Log RMSE:", rmse_log)
```

Log RMSE: 0.11079 on training set

```python
submission = pd.read_csv('data/sample_submission.csv')

y_test_pred = LR.predict(X_test)

submission['SalePrice'] = y_test_pred

submission.to_csv('data/Linear_submission.csv', index=False)
```

Recieved a log RMSE score of 11.94747 on test dataset.
